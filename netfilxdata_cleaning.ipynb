```python
import pandas as pd
df=pd.read_csv('netflix.csv')
df
```

```python
df.info()
```

```python
print(df.columns.tolist())
```

```python
df.drop(columns='rating')
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)
df
```

```python
df.duplicated().sum()
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.isna().sum()
#shows missing values
```

```python
#if we had na values
df.dropna(inplace=True)
df
```

```python
df.show_id.unique()
```

```python
df['show_id'] = df['show_id'].str.replace('s', '')
df
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df

```

```python
df.to_csv('netflix_cleaned.csv',index='False')
```
